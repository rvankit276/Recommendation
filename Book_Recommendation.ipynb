```python
number_of_ratings = book_ratings.groupby('book_title')['book_rating'].count().reset_index()
```

```python
final_rating.drop_duplicates(['user_id','book_title'],inplace=True)
```

```python
final_rating.isnull().sum()
```

### Item to Item Collaborative Filtering

```python
final_rating_pivot = final_rating.pivot(columns = 'user_id',index='book_title',values='book_rating')
```

```python
final_rating_pivot
```

```python
final_rating_pivot.fillna(0,inplace=True)
```

```python
from scipy.sparse import csr_matrix
final_rating_sparse = csr_matrix(final_rating_pivot)
```

```python
final_rating_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(final_rating_sparse)
```

```python
final_rating_pivot.iloc[224,:]
```

```python
distance , recommendations = model.kneighbors(final_rating_pivot.iloc[224,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
recommendations
```

```python
for i in range(len(recommendations)):
    print(final_rating_pivot.index[recommendations[i]])
```

### Testing Modle

```python
def book_recommendation(book_title):
    book_index = np.where(final_rating_pivot.index == book_title)[0][0]
    distance , recommendations = model.kneighbors(final_rating_pivot.iloc[book_index,:].values.reshape(1,-1),n_neighbors=6)
    books = final_rating_pivot.index[recommendations[0]]
    print(f"You searched '{book_title}'\n")
    print("The suggestion books are:\n")
    for book in books:
        if book != book_title:
            print(book)
    
```

```python
book_name = "Life of Pi"
book_recommendation(book_name)
```

```python
def recommend_for_user(user_id):
    try:
        user_index = np.where(final_rating_pivot_u.index == user_id)[0][0]
    except IndexError:
        print(f"User{user_id} not found.")
        return
    distance, neighbors = model1.kneighbors(final_rating_pivot_u.iloc[user_index,:].values.reshape(1,-1),n_neighbors = 6)
    print(f"Target User: {user_id}\n")
    print("Most similar users:")
    for i, neighbor_idx in enumerate(neighbors[0]):
        neighbor_user_id = final_rating_pivot_u.index[neighbor_idx]
        if neighbor_user_id != user_id:
            print(f"  {neighbor_user_id} (distance: {distance[0][i]:.2f})")
    target_user_books = set(final_rating_pivot_u.columns[final_rating_pivot_u.iloc[user_index]>0])
    similar_user_books = set()
    for neighbors_idx in neighbors[0]:
        similar_user_books.update(final_rating_pivot_u.columns[final_rating_pivot_u.iloc[neighbors_idx]>=8])
    recommended_books = similar_user_books - target_user_books
    print("\nRecommended books based on similar users:")
    for book in list(recommended_books)[:10]:
        print(f"  {book}")
```

```python
recommend_for_user(254)
```
